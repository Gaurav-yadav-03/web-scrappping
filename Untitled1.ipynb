```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Set up the WebDriver (make sure to have the appropriate driver, e.g., chromedriver for Chrome)
service = Service("C:\\Users\\igaur\\Downloads\\chromedriver-win32\\chromedriver-win32\\chromedriver.exe")
  # Correct path to chromedriver.exe
driver = webdriver.Chrome(service=service)

# Step 1: Open the website
driver.get("https://agriexchange.apeda.gov.in/Market/spsmeasures/Index#")

# Step 2: Find the select element and choose India
try:
    # Wait until the dropdown is clickable
    WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CLASS_NAME, 'floating-select.mySelecta'))
    )
    select_element = driver.find_element(By.CLASS_NAME, 'floating-select.mySelecta')
    select_element.click()

    # Choose India from the dropdown
    india_option = driver.find_element(By.XPATH, "//option[contains(text(), 'India')]")
    india_option.click()

    # Step 3: Click the search button
    search_button = driver.find_element(By.CLASS_NAME, 'srchbtn')
    search_button.click()

    # Step 4: Wait for the table to load
    WebDriverWait(driver, 20).until(
        EC.visibility_of_element_located((By.ID, 'tblSpsData'))
    )

    # Step 5: Extract the table data
    table = driver.find_element(By.ID, 'tblSpsData')

    # Extract rows from the table
    rows = table.find_elements(By.TAG_NAME, 'tr')

    # Prepare the data for the DataFrame
    table_data = []
    for row in rows:
        cols = row.find_elements(By.TAG_NAME, 'td')
        cols = [col.text for col in cols]
        table_data.append(cols)

    # Step 6: Save to Excel
    df = pd.DataFrame(table_data)
    df.to_excel("sps_data.xlsx", index=False)

    print("Data saved successfully to sps_data.xlsx")

finally:
    # Close the browser after the task is done
    driver.quit()

```
